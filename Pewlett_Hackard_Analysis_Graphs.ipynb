```python
import pandas as pd
import matplotlib.pyplot as plt
import dataframe_image as dfi
%matplotlib inline
```

#### Pie chart of titles retiring

```python
pie_data = "Data/current_retiring_titles.csv"
```

```python
pie_df = pd.read_csv(pie_data)
pie_df
```

```python
pie_df.info()
```

```python
labels = pie_df["title"]
values = pie_df["Count"]
textprops = {"fontsize":14}
Pie, ax = plt.subplots(figsize=(10, 8))
colorset = ["gold", "magenta", "lightblue", "green", "orange", "greenyellow","black"]
explode = [0,0,0.05,0.1,0.15,0.2, 0.3]

ax.pie(values, labels = labels, colors = colorset, 
       explode = explode, autopct='%0.3f%%', textprops=textprops, startangle=195)

Pie
```

```python
Pie.savefig("Images/current_retire_pie.png")
```

```python
CurrEmpDept_df = pd.read_csv("Data/curr_emp_dept.csv")
CurrEmpDept_df
```

```python
RetireDept_df=pd.read_csv("Data/retire_dept.csv")
RetireDept_df
```

```python
PercentRetire_df = pd.DataFrame({
    "Department":CurrEmpDept_df["dept_name"],
    "All Employees":CurrEmpDept_df["count"],
    "Retiring Employees":RetireDept_df["count"]
})
PercentRetire_df
```

```python
PercentRetire_df["Percent Retiring"]=(PercentRetire_df["Retiring Employees"]/PercentRetire_df["All Employees"]*100)
PercentRetire_df
```

```python
# Formatting
PercentRetire_df["Percent Retiring"]=PercentRetire_df["Percent Retiring"].map('{:.2f}%'.format)
```

```python
PercentRetire_df.info()
```

```python
PercentRetire_df["All Employees"]=PercentRetire_df["All Employees"].map('{:,}'.format)
PercentRetire_df["Retiring Employees"]=PercentRetire_df["Retiring Employees"].map('{:,}'.format)
```
